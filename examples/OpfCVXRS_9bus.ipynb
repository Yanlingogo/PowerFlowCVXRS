```julia
include("../src/CVXRS_OPF.jl")

network_data = PowerModels.parse_file("../cases/pglib_opf_case118_ieee.m");
network_data=opf_initialization(network_data, 0.0)
#[network_data["gen"][i]["ptc_factor"]=Int(network_data["bus"][string(gen["gen_bus"])]["bus_type"]==3) for (i,gen) in network_data["gen"]]
#[network_data["gen"][i]["ptc_factor"]=gen["gen_status"]*(gen["pmax"]-gen["pmin"]) for (i,gen) in network_data["gen"]]
[network_data["gen"][i]["ptc_factor"]=gen["gen_status"]*float((gen["pmax"]-gen["pmin"])>0) for (i,gen) in network_data["gen"]]

network_data=runpf(network_data)
test_runpf(network_data)

max_iter_SCRS=3
num_data=10000

gamma_save=[0,0.005,0.01,0.015,0.02,0.025,0.03,0.035,0.04,0.045,0.05]
cost_nominal_save=zeros(size(gamma_save,1))
cost_worst_save=zeros(size(gamma_save,1))
cost_worst_sampled=zeros(size(gamma_save,1))
count_bounded=zeros(size(gamma_save,1))
count_bounded_fixed=zeros(size(gamma_save,1))
count_normal=zeros(size(gamma_save,1))

sample_violation_normal=[]

#(Σ0,γ0)=network_data["uncertainty"]
#Σ0=Matrix(Diagonal((pl0.^2).*(pl0.>0.5).+1e-12))
#network_data["uncertainty"]=(Σ0,γ0)

for i=1:size(gamma_save,1)
    network_data_test=deepcopy(network_data)

    (Σ0,γ0)=network_data_test["uncertainty"]
    network_data_test["uncertainty"]=(Σ0,gamma_save[i])

    network_data_test, result_cvxr=scrs(network_data_test, max_iter_SCRS)
    cost_nominal_save[i]=result_cvxr["obj"][end]
    cost_worst_save[i]=result_cvxr["worst_cost"][end]
    
    (Σ0,γ0)=network_data_test["uncertainty"]
    network_data_test["uncertainty"]=(Σ0,gamma_save[i])
    pl_sampled,data_gencost_bounded,violation_count_bounded,sample_violation_bounded=sample_pf(network_data_test,num_data,1)
    cost_worst_sampled[i]=maximum(data_gencost_bounded)
    count_bounded[i]=violation_count_bounded

    (Σ0,γ0)=network_data_test["uncertainty"]
    network_data_test["uncertainty"]=(Σ0,gamma_save[3])
    pl_sampled,data_gencost_bounded,violation_count_normal,sample_violation_bounded=sample_pf(network_data_test,num_data,0)
    count_normal[i]=violation_count_normal
    
    (Σ0,γ0)=network_data_test["uncertainty"]
    network_data_test["uncertainty"]=(Σ0,gamma_save[11])
    pl_sampled,data_gencost,violation_count_bounded_fixed,sample_violation_normal=sample_pf(network_data_test,num_data,1)
    count_bounded_fixed[i]=violation_count_bounded_fixed
    
    println("Iteration ",i ,": ",violation_count_bounded_fixed,"  ",violation_count_normal,"  ",violation_count_bounded)
end

pyplot()
default(show = true)
plot(cost_nominal_save,levels=0,color=:blues)
plot!(cost_worst_sampled,levels=0,color=:red)
plot!(cost_worst_save,levels=0,color=:green)


```
